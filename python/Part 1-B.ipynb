## 2 Jupyter Notebooks

While working through this module, you will be employing a special environment for Python programming which has proven useful in the computational and data sciences. The overarching JupyterLab interface is a special environment, itself largely written in Python, but it doesn't just work with Python. It provides similar tools for interacting with other programming languages as well.

In particular, you will be using the Jupyter **notebook** environment for practicing/learning many of these tools. Therefore, you can use the GUI tools provided by the Launcher to create a new Python 3 notebook for practicing the exercises below. You can make new/separate notebooks for each module, or compile them together into a single notebook. That choice is largely up to you.

The key thing to understand is that there are two main methods for interacting with a notebook: **code cells** and **markdown cells**. There is a third type, raw cells, which we will not employ very often. When you first open a notebook, you will have a single code cell. It can be changed to a markdown cell by using the dropdown menu at the top of the notebook window. You can mix cell types throughout a notebook. The code cells are used for Python code, and the Markdown cells are used for (primarily) documentation.

Each cell consists of an input box where you can type in code/instructions. Once you have provided the desired instructions in a code cell, you can hold down the **shift** key and press **enter**. This will pass the code to the Python interpreter for execution. The output of the command/code will be displayed back just below the cell (not all commands generate output). You can go back and edit cells and rerun them to see how the output changes.
