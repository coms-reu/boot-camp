#### Listing files and directories

When you start a terminal, it will typically initiate the shell from the location of your **home directory**. This will also typically be clean of files and other things on a new account, but new directories and files can be added in this location. Directories can contain any number of files, and also other directories, so that directories have a hierarchical structure.

Either way, one of the first things you might want to do is ask the shell which directory you are currently working from, and then see what it contains.

**pwd** - print working directory

Directories are arranged hierarchically, and are separated by a single slash. This means that when I print my working directory, I will see not just the name of the directory in which I am working, but all of the directories which contain it as well.

```bash
pwd
```

**ls** - list files

You can take a look at the files in your working directory by using the ls command. Some files (beginning with a dot) will be hidden from view unless you add the -a or "all" option for the ls command.

```bash
ls
```

```bash
ls -a
```

You can see that the "." and ".." ".ipynb_checkpoints" directories were hidden from view unless the -a option was added. The ".ipynb_checkpoints" directory is just like any other directory, it is just hidden because it starts with a dot. However, the "." and ".." directories are special. A single dot is a way to refer to the current working directory (regardless of its actual name) while ".." is a way to refer to the directory in which the current directory resides (one level "up" in the directory hierarchy).

You can list files in these specific directories by using the ls command on them...

```bash
ls .ipynb_checkpoints
```

```bash
ls ..
```

#### Making and removing directories

**mkdir** - make directory

You can use this command to make a new directory.

**rmdir** - remove directory

You can use this command to remove a created directory. **Note** you can only remove a directory using this command once it no longer contains any files or additional directories. There are commands which can get around this limitation, but it's a good check to make sure you don't accidentally lose something. Why? Well, there is **no recycle bin** or **trash** folder as such when using the shell... when a file or directory is removed **it is permanently removed**. So, always be careful when removing things using the shell...

```bash
mkdir mystuff
```

```bash
ls
```

```bash
rmdir mystuff
```

```bash
ls
```

#### Changing to a different working directory

**cd** - change directory

Let's make a new directory and change the working directory to it...

```bash
mkdir mystuff
```

```bash
pwd
```

```bash
cd mystuff
```

```bash
pwd
```

```bash
ls -a
```

```bash
## Note that my previous working directory would be ".."
ls -a ..
```

```bash
cd ..
```

```bash
rmdir mystuff
```

You can see that we used the ".." with cd in order to return the working directory back down one level in the directory hierarchy so that we could remove the directory that we made. You can also use cd in a couple of other useful ways...

**cd** - change to your home directory
**cd ~** - change to your home directory ("~" can be used more generally to specify your home directory much like "." and ".." have special meaning).

Also, you can use **tab-completion** to save typing long file names...

Try making the mystuff directory, then type only "cd my". Finally, hit the tab key once (or twice).

You should see the directory name complete automatically. The shell can try to guess what directory or file name you intend to complete based on the letters you have typed so far. If there is more than one potential match, then the shell will complete the rest of the name as far as it can and then stop. Hitting tab a few more times will make it list the rest of the potential matches to try to provide you with additional help. Don't forget to practice this feature often, **tab-completion** is very useful.

**rm** - remove a file (permanently!)

Be careful when removing files...

#### Copying files

**cp original_file copy_of_original_file** - makes a copy of a file

Note that spaces are used to separate commands, options, file/directory names. Therefore, it is common to use the \_ character instead since it doesn't need to be treated in any special way compared to other letters, but closely resembles a space.

Again, **use caution** since this command will overwrite the last file name provided **without any warning**.

```bash
touch myfile.txt
```

```bash
ls
```

```bash
cp myfile.txt myotherfile.txt
```

```bash
ls
```

#### Moving files

**mv original_file new_location** - moves a file to a new location (possibly renaming the file)

You can use the command to either move a file to another directory, or just rename a file within the current directory (or both at the same time).

```bash
mkdir mystuff
```

```bash
ls
```

```bash
mv myfile.txt mystuff/
```

```bash
ls
```

```bash
ls mystuff
```

```bash
mv myotherfile.txt mystuff/newname.txt
```

```bash
ls
```

```bash
ls mystuff
```

```bash
## Dangerous - remove all files/directories under target (-r, recursive)...
rm -r mystuff
```

```bash
ls
```

#### Other useful skills...

**clear** - will clear the current terminal screen

**cat file** - will print the (text) file to the terminal screen

**less file** - will print the (text) file to the terminal screen one page at a time

Less even has a search feature (type forward slash followed by a word to find and press enter).

**head file** - will print the first (10) lines from the file to the screen

**tail file** - will print the last (10) lines from the file to the screen

These utilities can be very useful for parsing or exploring text-based files (which Python codes and many data files are).

### Wildcards

You can use the **wildcard** characters (\* or ?) to help you look for specific sets of files. The \* wildcard will match any name (list all files by itself), but can be combined with other specific filename details to limit the search. The ? wildcard will match any **single** character instead of a group.

ls (and most commands) can understand these wildcards. Here are some examples:

```bash
ls *.png
```

```bash
ls Part\ 1-?.ipynb
```

## 3. Setting Up Your Github Account

To manage code for your projects, a commonly-employed tool is **git** which tracks changes to code and files which are part of a project. Code takes time to develop, with mistakes made along the way and potentially multiple people working on the same code at the same time. This creates problems for project management. In some sense, **git** acts like a management tool for a project with multiple coders, similar to how Google Docs provides a way for people to work collaboratively on a document.

The shared cloud space we will be using is Github, which will host the project code on the internet so that it is accessible from anywhere. In order to connect to Github and make changes to these projects, you will need to set up a public/private keypair for secure shell (ssh) access to Github.

Using the **terminal**, you can utilize the following command:

**ssh-keygen** - generate a new public/private keypair

```bash
ssh-keygen
```

Due to some limitations in the notebook environment for bash, we won't show the rest of the output from the command above. However, you can just accept the default locations for the files and file names provided by just hitting enter until you get back the command prompt.

This command will create two files:
* `~/.ssh/id_rsa`
* `~/.ssh/id_rsa.pub`

Note these are files in a hidden directory (within your home directory).

You can view the public key file (`id_rsa.pub`) file by using the `cat` command at the terminal. You should never share the other file with anyone (this is private key that you should retain).

However, now that you have a keypair, you still need a GitHub account to connect it to:
[https://github.com/](https://github.com/)

Once you have your account, go to your account **Settings**:



Next, select **SSH and GPG Keys**:



and then add a **New SSH Key**:



You can fill in the **Title** with any unique name for your key that you would like, but **copy-paste** the contents of your `id_rsa.pub` file into the **Key** section of the form and them click **Add SSH Key**.




Once you have installed your key, then you can test it by **cloning** a github repository using the **SSH** protocol. The **HTTPS** protocol only allows you to read from a repository, but not write to one, and the **GitHub CLI** requires installation of additional software.

Here is a link to a (rather plain and empty) code repository in our GitHub organization: [https://github.com/coms-reu/testing](https://github.com/coms-reu/testing)

Just use the copy-paste button:



inside of the **Code** dialog to get a *clonable* link for the repository:



You can then go back to your **terminal** and clone the repository using the `git clone` command, for example:

```bash
git clone git@github.com:coms-reu/testing.git
```

If you get warnings to accept unknown keys from github.com, then you are fine because this happens the first time SSH connections are made to the github servers. These warnings will then go away and not come back (unless you delete some SSH-related settings and files). In the end, **if you see something similar to the above** then the clone was probably successful, but you can verify by changing into the new directoy created during the cloning process (`testing`) and see if there are some new files there:

```bash
cd testing
ls -l
```
